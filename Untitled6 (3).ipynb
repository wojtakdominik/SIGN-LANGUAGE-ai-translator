```python
from sklearn.model_selection import train_test_split
import os
```

```python
root_folder = '/content/asl_alphabet_data'
test_folder = os.path.join(root_folder,'asl_alphabet_test/asl_alphabet_test')
train_folder = os.path.join(root_folder,'asl_alphabet_train/asl_alphabet_train')
#creating validation folder
val_folder = os.path.join(root_folder, 'asl_alphabet_val/asl_alphabet_val')

if not os.path.exists(val_folder):
    os.makedirs(val_folder)
```

```python
def split_files(train,test,val,train_ratio=0.75,test_ratio=0.15,val_ratio=0.1):

  subfolders = [f.path for f in os.scandir(train) if f.is_dir()]
  for sub in subfolders:
    label = os.path.basename(sub)
    selected_files = [f for f in os.listdir(sub) if os.path.isfile(os.path.join(sub, f))]
    test_files, val_files = train_test_split(selected_files,train_size = test_ratio,test_size=val_ratio,random_state=2025)

    for file in test_files:
            shutil.move(os.path.join(sub, file), os.path.join(test, file))
    for file in val_files:
            shutil.move(os.path.join(sub, file), os.path.join(val, file))

    ##deleting moved files to adress data leakage
    #for file in test_files + val_files:
            #os.remove(os.path.join(sub, file))
```

```python
import json
from glob import glob
```

```python
def coco_train(train):
  train_coco_file = {
      "images": [],
      "annotations": [],
      "categories": []
  }
  categories = sorted([f.name for f in os.scandir(train) if f.is_dir()])

  for i,cat in enumerate(categories):
    train_coco_file["categories"].append({
        "id": i+1,
        "name": cat,
        "supercategory": "letter"
    })
  image_id, annotation_id = 0,0

  for category in categories:
        category_folder = os.path.join(train_folder, category)
        image_files = glob(os.path.join(category_folder, "*.jpg"))

        for image_file in image_files:
            image_id += 1
            img = cv2.imread(image_file)
            height, width, _ = img.shape

            train_coco_file["images"].append({
                "id": image_id,
                "file_name": os.path.basename(image_file),
                "height": height,
                "width": width
            })

            annotation_id += 1
            train_coco_file["annotations"].append({
                "id": annotation_id,
                "image_id": image_id,
                "category_id": categories.index(category) +1,
                "bbox": [0,0, width,height],
                "area": width * height,
                "iscrowd": 0
            })

  output_file = os.path.join(train_folder,'train_coco.json')
  with open(output_file, 'w') as f:
    json.dump(train_coco_file, f, indent=4)
```

```python
coco_train(train_folder)
```
